#Data Collection


```python
import pandas as pd
import yfinance as yf

# Define stock tickers and time range
tickers = ["AAPL", "TSLA"]
start_date = "2018-01-01"
end_date = "2023-12-31"

# Fetch historical stock data from Yahoo Finance
raw_data = yf.download(tickers, start=start_date, end=end_date)

# Save raw data before preprocessing
raw_data.to_csv("Raw_AAPL_TSLA_Stock_Data_2018_2023.csv")

# Display first few rows of the raw dataset
print("Raw Data Sample:")
print(raw_data.head())
```
